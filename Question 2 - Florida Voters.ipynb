```python
import re

informationofvotes=[] #Initializing lists required
stringlist=[]
solutionlist=[]

fp = open('FloridaVoters.html', 'r') #creating file pointer and accessing file in read mode

for line in fp:
    k=re.findall('<td>[A-Z.\s0-9,0-9-]+<\/td>', line)#extract data using Regex
    if k!=[]:
        informationofvotes.append(k)#Create a list with all the relevant data extracted from the website

fp.close()

for item in informationofvotes:
    item[0]=item[0].lstrip('<td>').rstrip('</td>')#formatting data extracted as required
    s=''.join(item[0])
    i=s.replace(',','')
    stringlist.append(i)

del stringlist[-5:] #remove last 5 items as they are the total votes

for i in range(0, len(stringlist), 6):
    solutionlist.append(stringlist[i:i + 6])#create list of lists having individual lists corresponding to one county from the main list

for i in range(0,len(solutionlist)):
    solutionlist[i][1]=int(solutionlist[i][1])
    solutionlist[i][2]=int(solutionlist[i][2])
    
solutionlist= sorted(solutionlist,key=lambda solutionlist:solutionlist[2]) #sort according to democratic votes

for row in solutionlist:
    print("{: >12} {: >12} {: >12}".format(*row)) #print in required format
```
